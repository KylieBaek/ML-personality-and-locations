```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
lc = pd.read_csv("./loc.csv",sep=',')
bf = pd.read_csv("./bff.csv",sep=',')
```

```python
lc =lc.T
bf =bf.T
```

```python
lc
```

```python
bf
```

```python
lc_s = lc.corr(method='spearman')
bf_s = bf.corr(method='spearman')
```

```python
lc_s
```

```python
bf_s
```

```python
x = []
y = []
for i in range(0,33):
    for j in range(0,33):
        y.append([bf_s[i][j]])
        x.append([lc_s[i][j]])
```

```python
plt.scatter(x,y)
plt.title("Location and BFF")
plt.xlabel("location coefficient")
plt.ylabel("bff coefficient")
plt.axis([-1.5,1.5,-1.5,1.5])
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X=pd.DataFrame(x),y=pd.DataFrame(y))
```

```python
plt.scatter(x,y)
plt.title("Location and BFF")
plt.xlabel("location coefficient")
plt.ylabel("bff coefficient")
plt.axis([-1.0,1.0,-1.0,1.0])
plt.plot(x,model.predict(pd.DataFrame(x)),color='red')
plt.show()
```
